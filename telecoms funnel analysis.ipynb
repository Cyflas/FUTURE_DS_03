```python
import pandas as pd

# firstly loading the data in
df = pd.read_csv("bank-full.csv", sep=";")

df['y'] = df['y'].str.capitalize()
df['contact'] = df['contact'].str.capitalize()
df['job'] = df['job'].str.title()

print("data imported and formatted. now, time to funnel!")

```

```python
#creating explicit binary flags for funnel stages
#stage 1: everyone in the dataset was contacted
df['stage_1_contacted'] = 1

#stage 2: engaged & stayed on call for over 2 minutes
df['stage_2_engaged'] = df['duration'].apply(lambda x: 1 if x > 120 else 0)

#stage 3: converted; customer agreement reached
df['stage_3_converted'] = df['y'].apply(lambda x: 1 if x == 'Yes' else 0)

#stage 4: checking numbers to make sure maths track
print("=== Funnel Stage Breakdown ===")
print(f"Total Leads Contacted: {df['stage_1_contacted'].sum()}")
print(f"Total Leads Engaged:   {df['stage_2_engaged'].sum()}")
print(f"Total Conversions:     {df['stage_3_converted'].sum()}")
```

```python
df.to_csv("bank_funnel_clean.csv", index=False)
print("\nthis was a great success! the file has been prepped and saved!")
```
